```python
import re
def clean_text(text):
  text=re.sub(r'[^a-zA-Z\s]','',text)
  text=re.sub(r'[x00-\x7F]+',' ',text)
  text=text.lower()
  return text
```

```python
STOP_WORDS = ['a', 'about', 'above', 'after', 'again', 'against', 'all', 'am', 'an', 'and', 'any', 'are', 'aren\'t', 'as', 'at', 'be', 'because', 'been', 'before', 'being', 'below', 'between', 'both', 'but', 'by', 'can', 'cannot', 'could', 'couldn\'t', 'did', 'didn\'t', 'do', 'does', 'doesn\'t', 'doing', 'don\'t', 'down', 'during', 'each', 'few', 'for', 'from', 'further', 'had', 'hadn\'t', 'has', 'hasn\'t', 'have', 'haven\'t', 'having', 'here', 'here\'s', 'hers', 'herself', 'how', 'how\'s', 'howsoever', 'i', 'i\'d', 'i\'ll', 'i\'m', 'i\'ve', 'if', 'in', 'into', 'is', 'isn\'t', 'is\'s', 'it', 'it\'s', 'itself', 'let', 'me', 'more', 'most', 'mustn\'t', 'my', 'myself', 'needn\'t', 'no', 'nor', 'not', 'of', 'off', 'on', 'once', 'only', 'or', 'other', 'ought', 'our', 'ours', 'ourselves', 'out', 'over', 'own', 'same', 'shan\'t', 'she', 'she\'s', 'should', 'shouldn\'t', 'so', 'some', 'such', 'than', 'that']
```

```python
def gen_freq(text):
  word_list=[]
  for tw_word in text.split():
    word_list.extend(tw_word)
    word_freq=pd.Series(word_list).value_counts()
    word_freq=word_freq.drop(STOP_WORDS,errors='ignore')
    return word_freq
```

```python
def any_neg(words):
  for word in words:
    if word in ['n','no','non','not'] or re.search(r"\wn't",word):
      return 1
    else:
      return 0
```
